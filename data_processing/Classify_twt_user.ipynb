```python
import os
ROOT = "/content/drive/MyDrive/json"
files = (os.listdir(ROOT))
```

```python
import pandas as pd
import json

df = {'id':[],'name':[],'username':[],'bio':[],'location':[],'url':[],'join_date':[],'join_time':[],
      'tweets':[],'following':[],'followers':[],'likes':[],'media':[],'private':[],'verified':[],
      'profile_image_url':[],'background_image':[]}
```

```python
import json
data = json.load(open(os.path.join(ROOT,files[0])))
```

```python
data[0]['author']['verified']
```

```python
data[0]['author'].keys()
```

```python
from tqdm import tqdm
for jfile in tqdm(files):
    data = json.load(open(os.path.join(ROOT,jfile)))
    for tweet in data:
        author = tweet['author']
        df['id'].append(author['id'])
        df['name'].append(author['name'])
        df['username'].append(author['username'])
        df['bio'].append(author['bio'])
        df['location'].append(author['location'])
        df['url'].append(author['profile_url'])
        jdt = author['date']
        df['join_date'].append(jdt[:jdt.find(' ')])
        df['join_time'].append(jdt[jdt.find(' ')+1:])
        df['tweets'].append(author['statuses_count'])
        df['following'].append(author['friends_count'])
        df['followers'].append(author['followers_count'])
        df['likes'].append(author['favourites_count'])
        df['media'].append(author['media_count'])
        df['private'].append(1 if author['protected'] else 0)
        df['verified'].append(1 if author['verified'] else 0)
        df['profile_image_url'].append(author['profile_image_url_https'])
        df['background_image'].append(author['profile_banner_url'])

```

```python
export_df = pd.DataFrame.from_dict(df)
len(export_df)
```

```python
new_df = export_df.drop_duplicates(subset=['id'])
```

```python
new_df.to_csv('users.csv',index=False)
```
